# Задача 40: Работать с файлом california_housing_train.csv, который находится в папке sample_data. Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population).

```python
df[(df['population'] >= 0) & (df['population'] < 501)]['median_house_value'].mean()
```

###### Из любопытства нашел и медианное значение и оно прилично отличается от среднего))

```python
df[(df['population'] >= 0) & (df['population'] < 501)]['median_house_value'].median()
```

# Задача 42: Узнать какая максимальная households в зоне минимального значения population
## Зону минимального значение количества людей взял 500% от минимального значения, так хотя бы больше выборка.

```python
df[df['population'] < df['population'].min() * 5]
```

```python
df[df['population'] < df['population'].min() * 5]['households'].max()
```

### Если вывести всю строку, тогда выводятся все строки с полученным значением.

```python
count = df[df['population'] < df['population'].min() * 5]['households'].max()
df[df['households'] == count]
```

```python
df[df['households'] == 8]
```
